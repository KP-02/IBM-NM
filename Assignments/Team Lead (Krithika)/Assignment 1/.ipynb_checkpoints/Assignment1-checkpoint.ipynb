```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the dataset
df = pd.read_csv('House Price India.csv')
```

```python
# 2.1 Univariate Analysis
# Histogram for living area
plt.figure(figsize=(8, 6))
plt.hist(df['living area'], bins=20, color='skyblue')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.title('Distribution of Living Area')
plt.show()
```

```python
# 2.2 Bi-Variate Analysis
# Scatter plot of living area vs. house price
plt.figure(figsize=(8, 6))
plt.scatter(df['living area'], df['Price'], color='darkblue')
plt.xlabel('Living Area')
plt.ylabel('House Price')
plt.title('Living Area vs. House Price')
plt.show()
```

```python
# 2.3 Multi-Variate Analysis
# Pairplot of selected features
selected_features = ['living area', 'lot area', 'number of bedrooms', 'number of bathrooms']
sns.pairplot(df[selected_features])
plt.show()
```

```python
# 3. Descriptive statistics
# Summary statistics for the dataset
summary_stats = df.describe()
print(summary_stats)
```

```python
# 4. Handling missing values
# Checking for missing values
missing_values = df.isnull().sum()
print(missing_values)

df['number of bathrooms'] = df['number of bathrooms'].fillna(df['number of bathrooms'].mean())
```
